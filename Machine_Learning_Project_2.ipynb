```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv ("titanic (1).csv",sep='\t')
df.head()
```

```python
plt.figure(figsize=(8, 12))
sns.heatmap(df.isnull(), cbar=False, cmap="cividis")
plt.title("Heatmap of Null Values (Before Removal)")
plt.show()
```

```python
df.drop("SibSp",axis = 1, inplace = True)
```

```python
df.head()
```

```python
df_clean = df.dropna()
```

```python
plt.figure(figsize=(8, 12))
sns.heatmap(df_clean.isnull(), cbar=False, cmap="cividis")
plt.title("Heatmap of Null Values (After Removal)")
plt.show()
```
