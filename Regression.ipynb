```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)

diabetes_X = diabetes_X[:, np.newaxis, 2]

diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]

regres= linear_model.LinearRegression()
regres.fit(diabetes_X_train, diabetes_y_train)
```

```python
diabetes_y_pred=regres.predict(diabetes_X_test)
print (diabetes_y_pred)
```

```python
print(f"Coefficient: {regres.coef_}")
print(f"Intercept: {regres.intercept_}")

print(f"Mean squared error: {mean_squared_error(diabetes_y_test, diabetes_y_pred)}")

print(f"Coefficient of determination (R2): {r2_score(diabetes_y_test, diabetes_y_pred)}")
```

```python
plt.scatter(diabetes_X_test, diabetes_y_test, color="black")
plt.plot(diabetes_X_test, diabetes_y_pred, color="red", linewidth=2)
plt.xticks(())
plt.yticks(())
plt.title("Linear Regression: Diabetes Dataset")
plt.legend()
plt.show()
```
