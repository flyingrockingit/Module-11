```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv ("US_Top_50_Universities_2026.csv")
```

```python
df.head (6)
```

```python
X = df['Research_Impact_Score'].values
y = df['Employment_Rate'].values
X = X.reshape(-1,1)
```

```python
X = (X - X.mean()) / X.std()
y = (y - y.mean()) / y.std()
```

```python
m = 0
b = 0
learning_rate = 0.01
epochs = 1000
n = len(X)
for i in range(epochs):
    y_pred = m * X.flatten() + b

    error = y - y_pred

    dm = (-2/n) * sum(X.flatten() * error)
    db = (-2/n) * sum(error)

    m = m - learning_rate * dm
    b = b - learning_rate * db

print("Best slope (m):", m)
print("Best intercept (b):", b)
```

```python
y_pred = m * X.flatten() + b
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel("Research Impact Score (Normalized)")
plt.ylabel("Employment Rate (Normalized)")
plt.title("Linear Regression using Gradient Descent")
plt.show()
```
