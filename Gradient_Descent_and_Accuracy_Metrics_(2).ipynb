```python
import numpy as np
from numpy import asarray, arange
from numpy.random import rand
from matplotlib import pyplot
```

```python
def objectives(x):
    return x**2.0

def derivative(x):
    return 2.0 * x
```

```python
def gradient_descent(objective, derivative, bounds, n_iter, step_size):
    solutions = []
    scores = []

    solution = bounds[:, 0] + rand(len(bounds)) * (bounds[:, 1] - bounds[:, 0])

    for i in range(n_iter):
        gradient = derivative(solution)
        solution = solution - step_size * gradient
        solution_eval = objective(solution)

        solutions.append(solution.copy())
        scores.append(solution_eval)

        print("%d f(%s) = %.5f" % (i, solution, solution_eval))

    return solutions, scores
```

```python
bounds = asarray([[-1.0, 1.0]])
n_iter = 30
step_size = 0.1
solutions, scores = gradient_descent(objectives, derivative, bounds, n_iter, step_size)
x = arange(bounds[0, 0], bounds[0, 1] + 0.01, 0.01)
results = objectives(x)
pyplot.plot(x, results)
pyplot.plot(solutions, scores, 'o', color='red')
pyplot.show()
```
