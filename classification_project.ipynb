```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv ("titanic (1).csv",sep='\t')
df.head(11)
```

```python
df.isnull().sum()
```

```python
df.drop('Cabin', axis=1, inplace=True)
df.fillna({
    'Age': df['Age'].mean(),
    'Embarked': df['Embarked'].mode()[0]
}, inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df,palette=['#d9534f', '#5cb85c'])
plt.title("Survived vs Dead")
plt.xticks([0,1], ['Dead', 'Survived'])
plt.xlabel("")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
female_data = df[df['Sex'] == 'female']
sns.set_style("darkgrid")
plt.figure(figsize=(6,4))
sns.countplot(x='Survived',data=female_data,palette=['#8172B3'])
plt.title("Female Survival Count", fontsize=16, weight='bold')
plt.xticks([0,1], ['Dead', 'Survived'])
plt.xlabel("")
plt.ylabel("Count")
sns.despine()
plt.tight_layout()
plt.show()
```

```python
male_data = df[df['Sex'] == 'male']
sns.set_style("darkgrid")
plt.figure(figsize=(6,4))
sns.countplot(x='Survived',data=male_data,palette=['#4C72B0'])
plt.title("Male Survival Count", fontsize=16, weight='bold')
plt.xticks([0,1], ['Dead', 'Survived'])
plt.xlabel("")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(6,4))
x = sns.countplot(x='Survived',hue='Pclass',data=df,palette=['#1f78b4', '#33a02c', '#e31a1c'])
plt.title("Survival Count by Passenger Class", fontsize=16, weight='bold')
plt.xticks([0,1], ['Dead', 'Survived'])
plt.xlabel("Survival Status")
plt.ylabel("Number of Passengers")
x.legend(title='Passenger Class', labels=['1st Class', '2nd Class', '3rd Class'])
plt.tight_layout()
plt.show()
```
